```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
LaLigaCupDf = pd.read_csv("Laliga.csv",skiprows=1)
LaLigaCupDf.tail()
```

```python
#Checking datatypes of our dataframe
LaLigaCupDf.info()
```

```python
LaLigaCup = LaLigaCupDf.replace('-',0)
```

```python
#Converting object datatypes to integers for calculations of our dataframe
LaLigaCup = LaLigaCup.astype({'Points':np.int64, 'GamesPlayed':np.int64, 'GamesWon':np.int64,
       'GamesDrawn':np.int64, 'GamesLost':np.int64, 'GoalsFor':np.int64, 'GoalsAgainst':np.int64, 'Champion':np.int64,
       'Runner-up':np.int64, 'Third':np.int64, 'Fourth':np.int64, 'Fifth':np.int64, 'Sixth':np.int64, 'T':np.int64})
LaLigaCup.info()
```

```python
LaLigaCup.tail()
```

```python
debut = pd.to_numeric(LaLigaCup.Debut.str.split('-').str[0])
teamsBetween_1930_1980 = LaLigaCup[debut.between(1930,1980)]
print(teamsBetween_1930_1980['Team'])
```

```python
LaLigaCup.groupby(['Points'])
```

```python
LaLigaCup_TopTeams_ByPoints = LaLigaCup.sort_values(by=['Points'],ascending=False)
print(LaLigaCup_TopTeams_ByPoints.iloc[0:5,1].values)
LaLigaCup_TopTeams_ByPoints.head()
```

```python
def Goal_diff_count(LaLigaCup):
    LaLigaCup['Goal_diff_count'] = LaLigaCup['GoalsFor'] - LaLigaCup['GoalsAgainst']
    return LaLigaCup[['Team','Goal_diff_count']]

team_goal_diff = Goal_diff_count(LaLigaCup)
team_goal_diff.sort_values(by='Goal_diff_count',ascending=False).iloc[0]
```

# Ques5 : Percentage of Winning = (GamesWon / GamesPlayed)*100

```python

LaLigaCup['Winning Percent'] = (LaLigaCup['GamesWon'] / LaLigaCup['GamesPlayed'] ) *100
LaLigaCup.fillna(value=0,inplace=True)
highestPercentageWinning =LaLigaCup.sort_values(by='Winning Percent',ascending=False).iloc[0:5,]
highestPercentageWinning
```

```python
highestPercentageWinning.loc[:,['Team','Winning Percent']]
```
